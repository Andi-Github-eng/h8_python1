```python
from pytictoc import TicToc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

t = TicToc()  # create instance of class

t.tic()  # Start timer
# Read csv data from specific folder
file01 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_20022020.csv'
file02 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_21022020.csv'
file03 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_22022020.csv'
file04 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_23022020.csv'
file05 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_24022020.csv'
file06 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_25022020.csv'
file07 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_26022020.csv'
file08 = 'C:/Users/Andi/H8_PFDS_DS_C1C2_3/3G_Profile.csv'
df1 = pd.read_csv(file01, header=None)
df2 = pd.read_csv(file02, header=None)
df3 = pd.read_csv(file03, header=None)
df4 = pd.read_csv(file04, header=None)
df5 = pd.read_csv(file05, header=None)
df6 = pd.read_csv(file06, header=None)
df7 = pd.read_csv(file07, header=None)
df8 = pd.read_csv(file08, na_values='\\N', usecols=['AREA', 'REGION', 'SALES_AREA', 'MICRO_CLUSTER', 'SITE_ID',
                                                    'RNC_ID', 'WBTS_ID', 'CELL_ID', 'Uarfcn', 'Band', 'Carrier',
                                                    'Sector_ID', 'Coverage'])
df1.dropna(how='any', inplace=True)
df2.dropna(how='any', inplace=True)
df3.dropna(how='any', inplace=True)
df4.dropna(how='any', inplace=True)
df5.dropna(how='any', inplace=True)
df6.dropna(how='any', inplace=True)
df7.dropna(how='any', inplace=True)
df8.fillna(value=0, inplace=True)
df8 = df8.astype({'RNC_ID': 'int16', 'WBTS_ID': 'int16',
                  'Sector_ID': 'int16', 'Uarfcn': 'int16', 'Band': 'int16'})
df8.rename(columns={'CELL_ID': 'WCEL_ID'}, inplace=True)
df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)
t.toc()  # Time elapsed since t.tic()

t.tic()  # Start timer
df.drop(columns=[0], inplace=True)
header = (['Date', 'Hour', 'RNC_ID', 'WBTS_ID', 'WCEL_ID',
           'Availability', 'HSDPA User', 'DL_power', 'UL_power', 'Code'])
df.rename(columns={1: 'Date', 2: 'Hour', 3: 'RNC_ID', 4: 'WBTS_ID', 5: 'WCEL_ID',
                   6: 'Availability', 7: 'HSDPA_User', 8: 'DL_power', 9: 'UL_power', 10: 'Code'}, inplace=True)

total = df.isnull().sum().sort_values(ascending=False)
percent = 100*(df.isnull().sum()/df.isnull().count()
               ).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
# print(missing_data)
t.toc()  # Time elapsed since t.tic()
t.tic()
df_combine = pd.merge(df, df8, how='left', on=['RNC_ID', 'WBTS_ID', 'WCEL_ID'])
#df_combine = df_combine.astype({'Sector_ID': 'int16', 'Uarfcn': 'int16', 'Band': 'int16'})
t.toc()
```
